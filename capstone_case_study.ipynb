```python
import pandas as pd
df = pd.read_csv('combined_data_2025.csv')

# Convert the columns
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])

# Check the data types again
print(df.info())
```

```python
df['ride_length'] = (df['ended_at'] - df['started_at']).dt.total_seconds() / 60
```

```python
df['ride_length']
```

```python
df = df[df['ride_length'] > 0]
```

```python
df
```

```python
df['day_of_week'] = df['started_at'].dt.day_name()
df['month'] = df['started_at'].dt.month
df['hour'] = df['started_at'].dt.hour
```

```python
df
```

```python
df.dropna(subset=['start_station_name', 'end_station_name'], inplace=True)
```

```python
df
```

```python
df= df[df['ride_length'] > 0]
```

```python
df
```

```python
avg_ride_length_by_member_type = df.groupby('member_casual')['ride_length'].mean()
print("Average Ride Length (minutes):\n", avg_ride_length_by_member_type)
```

```python
avg_ride_length_by_member_type = df.groupby('member_casual')['ride_length'].max()
print("Average Ride Length (minutes):\n", avg_ride_length_by_member_type)
```

```python
avg_ride_length_by_member_type = df.groupby('member_casual')['ride_length'].min()
print("Average Ride Length (minutes):\n", avg_ride_length_by_member_type)
```

```python
# Find the mode for 'day_of_week' for each rider type
most_popular_day_by_type = df.groupby('member_casual')['day_of_week'].agg(lambda x: x.mode()[0])
print(most_popular_day_by_type)
```

```python
rides_by_day = df.groupby(['member_casual', 'day_of_week']).size().unstack(fill_value=0)
print("\nRides by Day of the Week:\n", rides_by_day)
```

```python
# Top 10 start stations
top_start_stations = df.groupby('member_casual')['start_station_name'].value_counts().groupby(level=0).head(10)
print("\nTop Start Stations:\n", top_start_stations)

# Top 10 end stations
top_end_stations = df.groupby('member_casual')['end_station_name'].value_counts().groupby(level=0).head(10)
print("\nTop End Stations:\n", top_end_stations)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Make sure you have the df DataFrame loaded and cleaned before this step
# You can re-run the previous cleaning steps to be safe

# Group data to get average ride length by user type
avg_ride_length = df.groupby('member_casual')['ride_length'].mean().reset_index()

sns.barplot(x='member_casual', y='ride_length', data=avg_ride_length)
plt.title('Average Ride Length by Rider Type')
plt.xlabel('Rider Type')
plt.ylabel('Average Ride Length (minutes)')
plt.show()
```

```python
# Group data to get total rides by user type and day of week
rides_by_day = df.groupby(['member_casual', 'day_of_week']).size().unstack(fill_value=0)

# Reorder the columns for a logical flow (Monday to Sunday)
ordered_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
rides_by_day = rides_by_day[ordered_days]

# Create a line plot
rides_by_day.T.plot(kind='line', marker='o')
plt.title('Rides by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Rides')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

# 1. Average Ride Length
The average ride length for casual riders is significantly longer than for annual members.

Casual riders often use the bikes for longer, one-off trips for leisure or tourism.

Annual members tend to use bikes for shorter, purposeful trips like daily commutes.

# 2. Weekly Usage Patterns
The weekly usage patterns differ significantly between the two groups.

Casual riders show a clear spike in ridership on weekends (Saturday and Sunday), with lower activity during the week. This supports the idea that they are using the service for recreational purposes.

Annual members show consistent ridership throughout the weekdays, with a drop-off on weekends. This aligns with a commuter profile, using the bikes for daily travel to work or school.

# 3. Station Popularity
The popularity of stations likely reflects the different use cases.

Casual riders' most popular stations are often located near major tourist attractions, parks, and recreational areas.

Annual members' most popular stations are typically found in or between residential neighborhoods and business districts.

# Key Insights & Recommendations

# 1. Weekend-Only Pass

Casual riders mainly use bikes on weekends for long trips.

Offering a weekend-only membership matches their habits.

Converts one-time users into repeat riders, generating stable revenue.

Advertise at Tourist Spots

Casual riders often start trips at parks, museums, and tourist areas.

Ads placed at these spots (digital or physical) target the right people at the right time.

Avoids wasting budget on commuters who are already members.

# 2. Highlight Cost Savings

Casual riders pay more per trip compared to members.

Showing them exactly how much they could save with membership is a powerful motivator.

A cost-savings calculator or personalized ad can make the benefit clear.

# 3. Buy-One-Get-One Trip Offer

Casual riders often ride once and stop.

A BOGO promo encourages a second trip quickly, breaking the “one-and-done” habit.

Builds loyalty and increases chances of membership conversion.

Free Trip for New Members (First 2 Years)

Annual members get a free short trip in each of their first two years.

Reduces hesitation in signing up by showing immediate value.

Encourages renewals → lowers churn and strengthens loyalty.

# 4. Tiered Membership Rewards

Loyalty-based tiers encourage more rides.

Example: free trip for new members → rewards (discounts, cafe vouchers) for regular riders → premium perks (concert tickets, bike discounts) for super users.

Increases engagement, builds partnerships, and strengthens community feeling.

# ✨ Overall Conclusion:
Casual riders are motivated by fun, savings, and flexibility, not long-term commitments. These recommendations directly address their habits, build loyalty, and create multiple pathways to convert them into paying members while boosting Cyclistic’s long-term revenue.
