```python
import matplotlib.pyplot as plt

days = [1, 2, 3, 4]
sleeping = [4, 5, 8, 4]
dancing = [9, 8, 7, 8]
eating = [1, 2, 3, 4]
studying = [5, 8,4, 9]
explode = [0.4, 0, 0.7, 0]

activity_totals = [
    sum(sleeping),
    sum(dancing),
    sum(eating),
    sum(studying)
]

colors = ['#860AFA', '#0AFAE2', '#E2FA0A', '#FA520A']
labels = ['Sleeping', 'Dancing', 'Eating', 'Studying']

plt.figure(figsize=(10, 5))

# Line Plot (left)
plt.subplot(2, 2, 1)
plt.plot(days, sleeping, linestyle='--', linewidth=1, color='#0ACAFA', marker='^', label='Sleeping')
plt.plot(days, dancing, linestyle='-.', linewidth=1, color='#860AFA', marker='|', label='Dancing')
plt.title('Sleeping and Dancing Over Days')
plt.xlabel('Days')
plt.ylabel('Activity Time')
plt.legend()
plt.tight_layout()

# Pie Chart (right)
plt.subplot(2, 2, 2)
plt.pie(activity_totals, colors=colors, labels=labels, explode=explode, shadow=True, autopct='%1.1f%%')
plt.title('Total Time Spent on Activities')
plt.tight_layout()

plt.subplot(2, 2, 3)
plt.bar(labels, activity_totals, color=colors)
for i in range(len(days)):
    plt.text(i, activity_totals[i]+20, str(activity_totals[i]), ha='center')
plt.title('Activity Totals (Bar Graph)')
plt.ylabel('Total Time')

plt.subplot(2,2,4)
plt.stackplot(days,sleeping, dancing,eating,studying,colors=colors, alpha=0.8,labels=labels)
plt.title('activities')
plt.legend()
plt.tight_layout()


plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7,8]
y1 = [i*2 for i in x]
y2 = [i**2 for i in x]
y3 = [i**0.5 for i in x]
y4 = [10-i for i in x]

plt.figure(figsize=(10,8))  # Bigger size for clarity

# 2x2 layout
plt.subplot(2, 2, 1)
plt.plot(x, y1, linestyle='--', marker='^', color='#9E0AFA', label='y1', linewidth=2)
plt.title('Doubles')
plt.xlabel('x')
plt.ylabel('y1')

plt.subplot(2, 2, 2)
plt.plot(x, y2, linestyle='-', marker='^', color='#0AEEFA', label='y2', linewidth=2)
plt.title('Squares')
plt.xlabel('x')
plt.ylabel('y2')

plt.subplot(2, 2, 3)
plt.plot(x, y3, linestyle='-.', marker='^', color='#F6FA0A', label='y3', linewidth=2)
plt.title('Square Root')
plt.xlabel('x')
plt.ylabel('y3')

plt.subplot(2, 2, 4)
plt.plot(x, y4, linestyle=':', marker='^', color='#0AFA7E', label='y4', linewidth=2)
plt.title('Difference')
plt.xlabel('x')
plt.ylabel('y4')

plt.tight_layout()  # Auto-adjust spacing
plt.show()
```
